```python
"""
Rule-Based AI Chatbot
A simple chatbot that uses if-else logic to respond to predefined user inputs.
Runs in a continuous loop until the user says bye/exit/quit.
"""

import random
import datetime


"""
  RULE DEFINITIONS
  Each rule is a dict with:
    keywords  – list of trigger words/phrases
    responses – list of possible replies (one picked at random)
    tag       – label shown in the terminal for learning purposes
"""

RULES = [
    {
        "tag": "GREETING",
        "keywords": ["hello", "hi", "hey", "greetings", "howdy", "sup"],
        "responses": [
            "Hello there! Nice to meet you 👋",
            "Hey! How can I assist you today?",
            "Hi! Great to have you here.",
        ],
    },
    {
        "tag": "IDENTITY",
        "keywords": ["your name", "who are you", "what are you", "name"],
        "responses": [
            "I'm RuleBot — a simple rule-based AI chatbot!",
            "My name is RuleBot. I respond based on predefined if-else logic.",
        ],
    },
    {
        "tag": "STATUS",
        "keywords": ["how are you", "how r u", "how do you do", "feeling"],
        "responses": [
            "I'm doing great, thanks for asking!",
            "I'm just a program, but I'm running perfectly!",
            "All systems operational! How about you?",
        ],
    },
    {
        "tag": "JOKE",
        "keywords": ["joke", "funny", "laugh", "humor"],
        "responses": [
            "Why don't scientists trust atoms? Because they make up everything!",
            "Why did the programmer quit? Because he didn't get arrays!",
            "I told a joke about loops... and loops... and loops...",
        ],
    },
    {
        "tag": "WEATHER",
        "keywords": ["weather", "temperature", "forecast", "rain", "sunny"],
        "responses": [
            "I can't check live weather, but I hope it's sunny wherever you are! ☀️",
            "No weather API here — try checking your phone for the forecast!",
        ],
    },
    {
        "tag": "AGE",
        "keywords": ["how old", "your age", "age", "born"],
        "responses": [
            "I was born the moment someone ran this script!",
            "Age is just a number — and mine is 0, fresh off the press.",
        ],
    },
    {
        "tag": "THANKS",
        "keywords": ["thank", "thanks", "thank you", "thx", "appreciate"],
        "responses": [
            "You're welcome! 😊",
            "Happy to help!",
            "Anytime! That's what I'm here for.",
        ],
    },
    {
        "tag": "HELP",
        "keywords": ["help", "commands", "what can you do", "options"],
        "responses": [
            (
                "Here's what I understand:\n"
                "  • Greetings    – hello, hi, hey\n"
                "  • My identity  – who are you, your name\n"
                "  • My status    – how are you\n"
                "  • Jokes        – tell me a joke\n"
                "  • Time         – what time is it\n"
                "  • Date         – what is today's date\n"
                "  • Weather      – weather\n"
                "  • Thanks       – thank you\n"
                "  • Exit         – bye, exit, quit\n"
            )
        ],
    },
]

# Handled separately because they need live data
TIME_KEYWORDS  = ["time", "clock", "what time"]
DATE_KEYWORDS  = ["date", "today", "what day"]
EXIT_KEYWORDS  = ["bye", "goodbye", "exit", "quit", "see you", "farewell", "later"]



#Core Logic

def get_response(user_input: str) -> tuple[str, str]:
    """
    Match user input against rules using if-else logic.
    Returns (response_text, rule_tag).
    """
    text = user_input.lower().strip()


    if any(keyword in text for keyword in EXIT_KEYWORDS):
        farewell = random.choice([
            "Goodbye! It was nice chatting with you. 👋",
            "See you later! Come back anytime.",
            "Farewell! Take care.",
        ])
        return farewell, "EXIT"

    #time(dynamic)
    if any(keyword in text for keyword in TIME_KEYWORDS):
        now = datetime.datetime.now().strftime("%I:%M %p")
        return f"The current time is {now}.", "TIME"

    #date(dynamic)
    if any(keyword in text for keyword in DATE_KEYWORDS):
        today = datetime.datetime.now().strftime("%A, %B %d, %Y")
        return f"Today is {today}.", "DATE"


    for rule in RULES:
        if any(keyword in text for keyword in rule["keywords"]):
            response = random.choice(rule["responses"])
            return response, rule["tag"]


    fallback = random.choice([
        "I didn't quite catch that. Type 'help' to see what I can do.",
        "Hmm, I'm not sure how to respond to that. Try 'help'.",
        "I don't have a rule for that yet! Type 'help' for options.",
    ])
    return fallback, "DEFAULT"



#display helpers

def print_banner():

    print(" U L E - B A S E D   C H A T B O T")
    print("Simple AI using if-else logic")
    print("Type 'help' for commands  |  'bye' to exit")


def print_bot(message: str, tag: str):
    print(f"\n  [BOT | #{tag}]")
    for line in message.split("\n"):
        print(f"  {line}")
    print()



#main loop

def main():
    print_banner()
    print_bot("Hello! I'm RuleBot. How can I help you today?", "GREETING")

    while True:
        try:
            user_input = input("  You: ").strip()
        except (EOFError, KeyboardInterrupt):
            print_bot("Goodbye! (Session interrupted)", "EXIT")
            break

        if not user_input:
            continue

        response, tag = get_response(user_input)
        print_bot(response, tag)

        # Stop the loop on exit
        if tag == "EXIT":
            break


if __name__ == "__main__":
    main()
```
